# Regular Expression

>Here's an example of how to use regular expressions in Python to match and extract patterns from text:

```python
import re

# Example text
text = "Hello, my email address is example@email.com and my phone number is 123-456-7890."

# Pattern for matching email addresses
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'

# Pattern for matching phone numbers
phone_pattern = r'\b\d{3}-\d{3}-\d{4}\b'

# Find email addresses in the text
emails = re.findall(email_pattern, text)
print("Email addresses:")
for email in emails:
    print(email)

# Find phone numbers in the text
phone_numbers = re.findall(phone_pattern, text)
print("\nPhone numbers:")
for number in phone_numbers:
    print(number)
```

>In this example, we import the re module, which provides functions for working with regular expressions in Python. We define two patterns: email_pattern for matching email addresses and phone_pattern for matching phone numbers.

>We use the re.findall() function to find all occurrences of the patterns in the given text string. The results are stored in emails and phone_numbers lists, respectively.

>Finally, we iterate over the lists and print the extracted email addresses and phone numbers.

>Note that regular expressions can be more complex depending on the patterns you want to match. This is just a basic example to get you started.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
theta = np.linspace(0, 2*np.pi, 400)
x_boundary = 3 * np.cos(theta)
y_boundary = 2 * np.sin(theta)

plt.figure(figsize=(8,6))
plt.plot(x_boundary, y_boundary, label=r'$\frac{x^2}{9} + \frac{y^2}{4} = 1$', color='red')

plt.fill_between(x_boundary, y_boundary, 5, where=(x_boundary > -5), color='green', alpha=0.3)

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Define the boundary of the hyperbola (y^2 / 1 - x^2 / 9 = 1)
theta_hyperbola = np.linspace(0, 2*np.pi, 400)
x_boundary_hyperbola = 3 * np.cosh(theta_hyperbola)  # Semi-major axis of 3 (sqrt(9))
y_boundary_hyperbola = np.sinh(theta_hyperbola)  # Semi-minor axis of 1

# Create the plot for the domain outside the hyperbola
plt.figure(figsize=(8,6))
plt.plot(x_boundary_hyperbola, y_boundary_hyperbola, label=r'$\frac{y^2}{1} - \frac{x^2}{9} = 1$', color='blue')
plt.plot(-x_boundary_hyperbola, y_boundary_hyperbola, color='blue')
plt.plot(x_boundary_hyperbola, -y_boundary_hyperbola, color='blue')
plt.plot(-x_boundary_hyperbola, -y_boundary_hyperbola, color='blue')

# Shade the region outside the hyperbola
plt.fill_between(x_boundary_hyperbola, y_boundary_hyperbola, 5, where=(x_boundary_hyperbola > -5), color='orange', alpha=0.3)
plt.fill_between(-x_boundary_hyperbola, y_boundary_hyperbola, 5, where=(-x_boundary_hyperbola < 5), color='orange', alpha=0.3)

plt.title(r'Domain of $g(x, y) = \sqrt{-x^2 + 9y^2 - 9}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
# Define the boundary of the ellipse (x^2 + y^2 / 9 = 1)
theta_ellipse = np.linspace(0, 2*np.pi, 400)
x_boundary_ellipse = np.cos(theta_ellipse)  # Semi-major axis of 1 (sqrt(1))
y_boundary_ellipse = 3 * np.sin(theta_ellipse)  # Semi-minor axis of 3 (sqrt(9))

# Create the plot for the domain inside the ellipse
plt.figure(figsize=(8,6))
plt.plot(x_boundary_ellipse, y_boundary_ellipse, label=r'$x^2 + \frac{y^2}{9} = 1$', color='purple')

# Shade the region inside the ellipse
plt.fill_between(x_boundary_ellipse, y_boundary_ellipse, -3, where=(y_boundary_ellipse > -3), color='green', alpha=0.3)

plt.title(r'Domain of $g(x, y) = \sqrt{-9x^2 - y^2 + 9}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
# Define the boundary of the circle (x^2 + y^2 = 49)
theta_circle = np.linspace(0, 2*np.pi, 400)
x_boundary_circle = 7 * np.cos(theta_circle)  # Radius of 7
y_boundary_circle = 7 * np.sin(theta_circle)  # Radius of 7

# Create the plot for the domain outside the circle
plt.figure(figsize=(8,6))
plt.plot(x_boundary_circle, y_boundary_circle, label=r'$x^2 + y^2 = 49$', color='red')

# Shade the region outside the circle
plt.fill_between(x_boundary_circle, y_boundary_circle, 10, where=(x_boundary_circle > -10), color='blue', alpha=0.3)

plt.title(r'Domain of $g(x, y) = \sqrt{x^2 + y^2 - 49}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
# Define the boundary of the hyperbola (x^2 / 9 - y^2 / 4 = 1)
theta_hyperbola = np.linspace(0, 2*np.pi, 400)
x_boundary_hyperbola = 3 * np.cosh(theta_hyperbola)  # Semi-major axis of 3 (sqrt(9))
y_boundary_hyperbola = 2 * np.sinh(theta_hyperbola)  # Semi-minor axis of 2 (sqrt(4))

# Create the plot for the domain outside the hyperbola
plt.figure(figsize=(8,6))
plt.plot(x_boundary_hyperbola, y_boundary_hyperbola, label=r'$\frac{x^2}{9} - \frac{y^2}{4} = 1$', color='blue')
plt.plot(-x_boundary_hyperbola, y_boundary_hyperbola, color='blue')
plt.plot(x_boundary_hyperbola, -y_boundary_hyperbola, color='blue')
plt.plot(-x_boundary_hyperbola, -y_boundary_hyperbola, color='blue')

# Shade the region outside the hyperbola
plt.fill_between(x_boundary_hyperbola, y_boundary_hyperbola, 10, where=(x_boundary_hyperbola > -10), color='orange', alpha=0.3)
plt.fill_between(-x_boundary_hyperbola, y_boundary_hyperbola, 10, where=(-x_boundary_hyperbola < 10), color='orange', alpha=0.3)

plt.title(r'Domain of $g(x, y) = \frac{1}{\sqrt{4x^2 - 9y^2 - 36}}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
