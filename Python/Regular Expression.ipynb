# Regular Expression

>Here's an example of how to use regular expressions in Python to match and extract patterns from text:

```python
import re

# Example text
text = "Hello, my email address is example@email.com and my phone number is 123-456-7890."

# Pattern for matching email addresses
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'

# Pattern for matching phone numbers
phone_pattern = r'\b\d{3}-\d{3}-\d{4}\b'

# Find email addresses in the text
emails = re.findall(email_pattern, text)
print("Email addresses:")
for email in emails:
    print(email)

# Find phone numbers in the text
phone_numbers = re.findall(phone_pattern, text)
print("\nPhone numbers:")
for number in phone_numbers:
    print(number)
```

>In this example, we import the re module, which provides functions for working with regular expressions in Python. We define two patterns: email_pattern for matching email addresses and phone_pattern for matching phone numbers.

>We use the re.findall() function to find all occurrences of the patterns in the given text string. The results are stored in emails and phone_numbers lists, respectively.

>Finally, we iterate over the lists and print the extracted email addresses and phone numbers.

>Note that regular expressions can be more complex depending on the patterns you want to match. This is just a basic example to get you started.

```python
import random

def roll_dice():
    return random.randint(1, 6)

def simulate_game(rounds=10000000):
    A_win = 0
    B_win = 0
    Tie = 0
    results = []

    for round_num in range(1, rounds + 1):
        dice_a1 = roll_dice()
        dice_a2 = roll_dice()
        score_a = max(dice_a1, dice_a2)

        dice_b = roll_dice()
        score_b = dice_b + 1

        if score_a > score_b:
            winner = "A"
            A_win += 1
        elif score_a < score_b:
            winner = "B"
            B_win += 1
        else:
            winner = "Tie"
            Tie += 1

        results.append({
            "Round": round_num,
            "A (Max of Two Rolls)": score_a,
            "B (Roll + 1)": score_b,
            "Winner": winner
        })

    return results, A_win, B_win, Tie


if __name__ == "__main__":
    game_results, A, B, Tie = simulate_game()

    print("\nGame Results:\n")
    for result in game_results:
        print(f"Round {result['Round']}: A = {result['A (Max of Two Rolls)']}, "
              f"B = {result['B (Roll + 1)']}, Winner = {result['Winner']}")

    print("\nresults:")
    print(f"A wins: {A}")
    print(f"B wins: {B}")
    print(f"Ties: {Tie}")


```

```python
import random
import math

def simulate_darts(throws=1):
    square_side = 2
    circle_radius = 1
    hits_in_circle = 0

    for _ in range(throws):
        x = random.uniform(-circle_radius, circle_radius)
        y = random.uniform(-circle_radius, circle_radius)

        if math.sqrt(x**2 + y**2) <= circle_radius:
            hits_in_circle += 1

    return hits_in_circle, throws

if __name__ == '__main__':
    for _ in range(5):
        hits, total_throws = simulate_darts()
        print(f'-------------------')
        print(f"Total throws: {total_throws}")
        print(f"Hits inside circle: {hits}")
        print(f'Pi estimation: {4 * hits / total_throws}')
        print(f'Error: {abs(math.pi - 4 * hits / total_throws)}')
        print(f'-------------------')
```

```python
import numpy as np

def init_board(size=15):
    return np.zeros((size, size), dtype=int)

def print_board(board):
    symbols = {0: '.', 1: 'X', 2: 'O'}
    for row in board:
        print(' '.join(symbols[cell] for cell in row))

def check_winner(board, player):
    size = len(board)
    directions = [(0, 1), (1, 0), (1, 1), (1, -1)]

    for x in range(size):
        for y in range(size):
            if board[x][y] == player:
                for dx, dy in directions:
                    count = 0
                    for k in range(5):
                        nx, ny = x + k * dx, y + k * dy
                        if 0 <= nx < size and 0 <= ny < size and board[nx][ny] == player:
                            count += 1
                        else:
                            break
                    if count == 5:
                        return True
    return False

def play_gomoku():
    board = init_board()
    current_player = 1
    turn_count = 0

    while True:
        print_board(board)
        x, y = map(int, input(f"Player {current_player} (x y): ").split())

        if board[x, y] != 0:
            print("Position already taken. Try again.")
            continue

        board[x, y] = current_player
        turn_count += 1

        if check_winner(board, current_player):
            print_board(board)
            print(f"Player {current_player} wins!")
            break

        if turn_count == len(board) * len(board):
            print("It's a draw!")
            break

        current_player = 3 - current_player

if __name__ == "__main__":
    play_gomoku()
```

```python

# Algorithm to approximate Pi using the Levaillant method
def compute_pi():
    k = 1000000  # Number of iterations
    n = 3  # Number of terms in the formula
    approx = 0.0

    for i in range(1, k):
        term = 4.0 / (2 * i - 1)
        approx += term / ((2 * i) ** n)

    print("Approximation of Pi:")
    print(approx)


# Run the calculation
compute_pi()
```

```python
def fibonacci_recursive(n, fib=[0, 1]):
    if n <= len(fib):
        return fib[:n]
    fib.append(fib[-1] + fib[-2])
    return fibonacci_recursive(n, fib)

print(fibonacci_recursive(100))
```

```python
def fibonacci_sequence(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib

n = 10  # Change this value to generate more Fibonacci numbers
print(f"{fibonacci_sequence(n)}")
```

```python
def collatz(n):
    if n == 1:
        return [1]
    elif n % 2 == 0:
        return [n] + collatz(n // 2)
    else:
        return [n] + collatz(3 * n + 1)

sequence = collatz(1.1)

print(sequence)
print(len(sequence))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def run_sir_model(S_init, I_init, R_init, alpha, gamma, days):
    S, I, R = S_init, I_init, R_init
    S_hist, I_hist, R_hist = [], [], []
    
    for day in range(days):
        S_hist.append(S)
        I_hist.append(I)
        R_hist.append(R)
        
        delta_S = -alpha * S * I
        delta_I = alpha * S * I - gamma * I
        delta_R = gamma * I
        
        S += delta_S
        I += delta_I
        R += delta_R
        
        S, I, R = max(S, 0), max(I, 0), max(R, 0)

    return np.array(S_hist), np.array(I_hist), np.array(R_hist)

# Model parameters
alpha = 0.000218  # Infection rate
gamma = 0.44      # Recovery rate
days = 40         # Simulation time

# Case 1: No vaccination (Full susceptible population)
S1, I1, R1 = run_sir_model(9150, 1, 0, alpha, gamma, days)
peak_I1 = np.max(I1)
peak_day1 = np.argmax(I1)

# Case 2: 50% vaccinated (Reducing initial susceptible population)
S2, I2, R2 = run_sir_model(4575, 1, 0, alpha, gamma, days)
peak_I2 = np.max(I2)
peak_day2 = np.argmax(I2)

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(I1, label="Infected (No Vaccination)", color='red')
plt.plot(I2, label="Infected (50% Vaccinated)", color='blue', linestyle='dashed')
plt.xlabel("Days")
plt.ylabel("Number of Infected Individuals")
plt.title("SIR Model - Flu Outbreak at UC Merced")
plt.legend()
plt.grid(True)
plt.show()

import ace_tools as tools
import pandas as pd

# Display results in table format
data = pd.DataFrame({
    "Scenario": ["No Vaccination", "50% Vaccinated"],
    "Peak Infected": [peak_I1, peak_I2],
    "Peak Day": [peak_day1, peak_day2]
})
tools.display_dataframe_to_user(name="SIR Model Results", dataframe=data)
```
